```python
print("ПРОПУЩЕННЫЕ ЗНАЧЕНИЯ:\n")
missing_data = pd.DataFrame({
    'Тип_данных': df.dtypes,
    'Пропусков': df.isnull().sum(),
    '% пропусков': (df.isnull().sum() / len(df) * 100).round(2)
})
missing_data = missing_data[missing_data['Пропусков'] > 0].sort_values('Пропусков', ascending=False)
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

basic_stats = []
for col in numeric_cols:
    col_data = df[col].dropna()
    
    mode_values = col_data.mode()
    
    basic_stats.append({
        'Признак': col,
        'Минимум': col_data.min(),
        'Максимум': col_data.max(),
        'Среднее (mean)': col_data.mean(),
        'Медиана (median)': col_data.median(),
        'Мода (mode)': mode_values[0],
    })

basic_df = pd.DataFrame(basic_stats).round(2)
display(basic_df)
```

```python
percentiles = [5, 25, 50, 75, 95]
percentile_data = []

for col in numeric_cols:
    col_data = df[col].dropna()
    percentiles_dict = {'Признак': col}
    
    for p in percentiles:
        percentiles_dict[f'{p}%'] = np.percentile(col_data, p)
    
    percentile_data.append(percentiles_dict)

percentile_df = pd.DataFrame(percentile_data).round(2)
display(percentile_df)
```

```python
advanced_stats = []
for col in numeric_cols:
    col_data = df[col].dropna()
    
    advanced_stats.append({
        'Признак': col,
        'Дисперсия (variance)': col_data.var(),
        'Асимметрия (skewness)': col_data.skew(),
        'Эксцесс (kurtois)': col_data.kurtosis(),
    })

advanced_df = pd.DataFrame(advanced_stats).round(2)
display(advanced_df)
```

Судя по этим параметрам: 
1)Разброс в количестве денег в транзакциях большой, много малых хначений, больше редких, и пик очень острый
2)Возраст не так сильно разбросан, почти нормальное распределение, плоский пик
3)Длительность транзакции варьируется, чуть больше малых значений, около-нормальный пик
4)Количество попыток ввода почти везде равняется среднему(1), очень много малых значений, очень отсрый пик
5)Огромный разброс баланса, немного больше малых значений, пологий пик

```python
categories = []
for col in categorical_cols:
    col_data = df[col].unique()
    categories.append({
        'Признак' : col, 
        'Количество' : len(col_data)
    })
categories_df = pd.DataFrame(categories)
display(categories_df)
```

```python
low_count_categories_list = categories_df.iloc[:, 1]
low_count_rows = low_count_categories_list < 10
low_count_rows = low_count_categories_list[low_count_rows].index.tolist()
high_count_categories_df = categories_df.drop(index=low_count_rows)
low_count_categories_df = categories_df.loc[categories_df.index.isin(low_count_rows)]

low_count_mask = df.columns.isin(low_count_categories_df.iloc[:, 0].tolist())
high_count_mask = df.columns.isin(high_count_categories_df.iloc[:, 0].tolist())

# display(low_count_mask)
# display(high_count_mask)
```

```python
onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
low_cardinality_encoded = onehot_encoder.fit_transform(df.loc[:, low_count_mask])
```

```python
high_count_dicts = df.loc[:,high_count_mask].to_dict('records')
hasher = FeatureHasher(n_features=6, input_type='dict', alternate_sign=False)
high_cardinality_encoded = hasher.fit_transform(high_count_dicts).toarray()

numeric_data = df[numeric_cols].values

# display(low_cardinality_encoded)
display(high_cardinality_encoded)
# display(numeric_data)


df_encoded = np.hstack([low_cardinality_encoded, 
                        high_cardinality_encoded, 
                        numeric_data])
df_encoded = pd.DataFrame(df_encoded)

display(df_encoded)
display(df)
```
